### What is the Iris dataset?

A collection of data about iris flowers used for statistical classification and machine learning practice. The Iris dataset is a famous multivariate dataset introduced by Ronald Fisher in 1936.



### Types of Iris flowers

``1. Iris setosa - Small flower``

• Petals are shorter compared to other species

• Easily separable from the others

``2. Iris versicolor - Medium-sized flower``

• Petal measurements are between setosa and virginica

``3. Iris virginica - Largest flower among the three``

• Longest petals and sepals



### Each sample has four features (measurements):

⁃ Sepal length

⁃ Sepal width

⁃ Petal length

⁃ Petal width



# About the dataset


| **Column Name**   | **Definition**           | **Units**        | **Details**                                                                                                          |
| ----------------- | ------------------------ | ---------------- | -------------------------------------------------------------------------------------------------------------------- |
| **Id**            | Row number or identifier | None             | A unique number for each sample. Some datasets include this as the first column.                                     |
| **SepalLengthCm** | Length of the sepal      | Centimetres (cm) | Sepal is the outer part of the flower that protects the petals when in bud. This measures how long it is.            |
| **SepalWidthCm**  | Width of the sepal       | Centimetres (cm) | How wide the sepal is at its broadest part.                                                                          |
| **PetalLengthCm** | Length of the petal      | Centimetres (cm) | Petal is the colourful part of the flower. This measures its length.                                                 |
| **PetalWidthCm**  | Width of the petal       | Centimetres (cm) | How wide the petal is at its broadest part.                                                                          |
| **Species**       | Type of Iris flower      | Category         | The class label indicating which species the flower belongs to: **Iris-setosa, Iris-versicolor, or Iris-virginica**. |

### boxplot

A box plot (also called box-and-whisker plot) is a graph that shows the distribution of numerical data through five summary statistics:

◦ Minimum

◦ First quartile (Q1)

◦ Median (Q2)

◦ Third quartile (Q3)

◦ Maximum

```python
fig=sns.boxplot(x='Species',y='PetalLengthCm',data=iris,hue='Species',order=['Iris-virginica','Iris-versicolor','Iris-setosa'])
```

| Part                                                           | Meaning                                                                                                        |
| -------------------------------------------------------------- | -------------------------------------------------------------------------------------------------------------- |

| **`x='Species'`**                                              | Puts **Species categories on the x-axis**.                                                                     |
| **`y='PetalLengthCm'`**                                        | Puts **PetalLengthCm values on the y-axis**.                                                                   |
| **`data=iris`**                                                | Uses the **iris DataFrame** as the data source.                                                                |
| **`hue='Species'`**                                            | Colours each box by **Species category**, adding a legend.                                                     |
| **`order=['Iris-virginica','Iris-versicolor','Iris-setosa']`** | Specifies the **order in which Species appear on the x-axis** (Virginica first, then Versicolor, then Setosa). |


```python
iris.boxplot(by="Species", figsize=(12, 6))
```

### violin plot

A violin plot is a combination of a box plot and a KDE (Kernel Density Estimate) plot.

It is used to visualize the distribution of data and its probability distribution.This chart is a combination of a Box Plot and a Density Plot that is rotated and placed on each side, to show the distribution shape of the data. The thick black bar in the centre represents the interquartile range, the thin black line extended from it represents the 95% confidence intervals, and the white dot is the median.Box Plots are limited in their display of the data, as their visual simplicity tends to hide significant details about how values in the data are distributed

It shows:

• Summary statistics (median, quartiles) like a box plot

• Distribution shape (density) like a rotated density plot on both sides

```python
fig=sns.violinplot(x='Species',y='SepalLengthCm',data=iris,hue='Species')
```

```python
plt.figure(figsize=(15,10))
plt.subplot(2,2,1)
sns.violinplot(x='Species',y='PetalLengthCm',data=iris,palette='pastel')
plt.subplot(2,2,2)
sns.violinplot(x='Species',y='PetalWidthCm',data=iris,palette='summer')
plt.subplot(2,2,3)
sns.violinplot(x='Species',y='SepalLengthCm',data=iris,palette='winter')
plt.subplot(2,2,4)
sns.violinplot(x='Species',y='SepalWidthCm',data=iris,palette='Set2')
```

| Part                              | Meaning                                                                                   |
| --------------------------------- | ----------------------------------------------------------------------------------------- |
| **`plt.figure(figsize=(15,10))`** | Creates a **figure window** with size **15 inches wide and 10 inches high**.              |
| **`plt.subplot(2,2,1)`**          | Creates the **first subplot** in a **2x2 grid** (top-left).                               |
| **`sns.violinplot(...)`**         | Plots a **violin plot** for the given feature and species, with specified colour palette. |
| **`plt.subplot(2,2,2)`**          | Creates the **second subplot** (top-right).                                               |
| **`plt.subplot(2,2,3)`**          | Creates the **third subplot** (bottom-left).                                              |
| **`plt.subplot(2,2,4)`**          | Creates the **fourth subplot** (bottom-right).                                            |
| **`palette='...'`**               | Sets the **colour palette** for each plot (**pastel, summer, winter, Set2**).             |

### Heatmap

A heatmap is a graphical representation of data where values are shown as colours.

In data analysis, it is often used to show correlation matrices or tables, with colours indicating high or low values.



```python
numeric_iris = iris.select_dtypes(include=['float64', 'int64'])
plt.figure(figsize=(10,7))
sns.heatmap(numeric_iris.corr(),
            annot=True,
            cmap='cubehelix',
            linewidths=1,
            linecolor='k',
            square=True,
            vmin=-1,
            vmax=1,
            cbar_kws={"orientation": "vertical"},
            cbar=True)
```

| Line / Argument                                                       | Explanation                                                                                                   |
| --------------------------------------------------------------------- | ------------------------------------------------------------------------------------------------------------- |
| **`numeric_iris = iris.select_dtypes(include=['float64', 'int64'])`** | Creates a new DataFrame `numeric_iris` containing **only numeric columns** (float and int types) from `iris`. |
| **`plt.figure(figsize=(10,7))`**                                      | Creates a **figure window** with size **10 inches wide, 7 inches high**.                                      |
| **`sns.heatmap(...)`**                                                | Draws a **heatmap** showing correlations between numeric features.                                            |
| **`numeric_iris.corr()`**                                             | Calculates the **correlation matrix** for numeric\_iris DataFrame.                                            |
| **`annot=True`**                                                      | **Displays correlation values** inside each heatmap cell.                                                     |
| **`cmap='cubehelix'`**                                                | Uses the **cubehelix colour palette** for the heatmap.                                                        |
| **`linewidths=1`**                                                    | Adds **1-pixel wide lines** between cells.                                                                    |
| **`linecolor='k'`**                                                   | Sets **line colour to black** between cells.                                                                  |
| **`square=True`**                                                     | Makes **each cell square-shaped** for neatness.                                                               |
| **`vmin=-1`, `vmax=1`**                                               | Sets the **colour scale range from -1 to +1** (standard correlation range).                                   |
| **`cbar_kws={"orientation": "vertical"}`**                            | Sets the **colour bar orientation to vertical**.                                                              |
| **`cbar=True`**                                                       | Displays the **colour bar** to indicate correlation scale.                                                    |

### swarm plot

A swarm plot is a type of scatter plot that shows individual data points for a categorical variable, while adjusting them along the categorical axis so they don’t overlap.

It looks a bit like a friendly swarm of bees buzzing about their hive. More importantly, each data point is clearly visible and no data are obscured by overplotting.A beeswarm plot improves upon the random jittering approach to move data points the minimum distance away from one another to avoid overlays. The result is a plot where you can see each distinct data point, like shown in below plot


```python
plt.figure(figsize=(10,7))
sns.set(style="darkgrid")
fig = sns.swarmplot(x="Species", y="PetalLengthCm", data=iris,hue='Species')
```

| Line                             | Meaning                                                                                             |
| -------------------------------- | --------------------------------------------------------------------------------------------------- |
| **`plt.figure(figsize=(10,7))`** | Creates a new figure with **width 10 inches** and **height 7 inches**                               |
| **`sns.set(style="darkgrid")`**  | Sets the **Seaborn plot style** to ‘darkgrid’, adding a grey grid background for better readability |
| **`sns.swarmplot(...)`**         | Creates a **swarm plot**                                                                            |
| **`x="Species"`**                | x-axis shows **Species categories** (Setosa, Versicolor, Virginica)                                 |
| **`y="PetalLengthCm"`**          | y-axis shows **PetalLengthCm values**                                                               |
| **`data=iris`**                  | Uses the **iris DataFrame**                                                                         |
| **`hue='Species'`**              | Colours the points **based on Species**, giving different colours for each class                    |


```python
plt.figure(figsize=(10,7))
sns.set(style="whitegrid")
fig = sns.swarmplot(x="Species", y="PetalLengthCm", data=iris, color="white")
ax = sns.violinplot(x="Species", y="PetalLengthCm", data=iris, hue='Species')
```
