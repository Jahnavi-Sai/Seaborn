# 1️. Introduction to EDA

Exploratory Data Analysis (EDA) is the critical first step in any data analysis or machine learning project. It involves examining datasets to summarise their main characteristics, often using visual methods.

### Key points about EDA:

⋗ Helps understand data structure, patterns, and relationships

⋗ Identifies missing values, outliers, and anomalies

⋗ Provides insights for feature selection, engineering, and modelling decisions

⋗ Facilitates hypothesis generation and testing

In this project, we will perform a comprehensive EDA on the Heart Disease dataset to uncover meaningful insights about patient health parameters and their relation to heart disease outcomes.

---

# 2️. Objectives of EDA

The main objectives of Exploratory Data Analysis (EDA) are to:

⤠ ``Understand the dataset structure and variables``- Gain familiarity with features, data types, and distributions.

⤠ ``Detect missing values and anomalies``- Identify incomplete or inconsistent data entries.

⤠ ``Identify outliers``- Find unusual data points that may impact analysis or model performance.

⤠ ``Discover patterns and relationships``- Understand how features relate to each other and to the target variable.

⤠ ``Generate hypotheses for modelling``- Develop insights that guide feature engineering and predictive analysis.

⤠ ``Select appropriate visualisation techniques``- Effectively communicate data insights using plots and charts.

---

# 3️. Types of EDA

Exploratory Data Analysis (EDA) can be classified along two dimensions:

``1. Graphical vs Non-Graphical :``

⊹ Graphical methods use visualisations to summarise data.

⊹ Non-graphical methods use summary statistics and numerical measures.

``2. Univariate vs Multivariate (including Bivariate) :``

⊹ Univariate focuses on a single variable.

⊹ Bivariate/Multivariate focuses on relationships between two or more variables.

### The Four Types of EDA

``⟡ i. Univariate Non-Graphical EDA``

Focuses on summary statistics such as mean, median, standard deviation, and outlier detection for a single variable.

``⟡ ii. Multivariate Non-Graphical EDA``

Uses cross-tabulations or correlation matrices to understand relationships between multiple variables without visualisation.

``⟡ iii. Univariate Graphical EDA``

Uses histograms, boxplots, and bar charts to visualise the distribution and spread of a single variable, providing insights that summary statistics alone may miss.

``⟡ iv. Multivariate Graphical EDA``

Uses plots to analyse relationships between two or more variables. Common techniques include:

⁃ Side-by-side boxplots

⁃ Scatterplots

⁃ Heatmaps

⁃ 3D surface plots

# 10. Handling Missing Values

### How Pandas Represents Missing Data

In Pandas, missing data is represented by:

1. None:

A Python singleton object used to indicate missing data in general Python code.

2. NaN (Not a Number):

A special floating-point value (IEEE standard) used by Pandas to represent missing numerical data.

◼ Detecting Missing Values

Pandas provides two main functions to detect missing data:

⦂ isnull() – returns True for missing values, otherwise False.

⦂ notnull() – returns True for non-missing values, otherwise False.

### Useful Commands to Check Missing Values
| **Command**                | **Description**                                                                          |
| -------------------------- | ---------------------------------------------------------------------------------------- |
| `df.isnull()`              | Returns a DataFrame of boolean values indicating missing data (True for missing).        |
| `df.isnull().sum()`        | Total number of missing values **in each column**.                                       |
| `df.isnull().sum().sum()`  | Total number of missing values **in the entire DataFrame**.                              |
| `df.isnull().mean()`       | Proportion (percentage) of missing values **in each column**.                            |
| `df.isnull().any()`        | Checks **which columns have missing values**, returns True/False for each column.        |
| `df.isnull().any().any()`  | Returns a single True/False indicating **if any missing value exists** in the DataFrame. |
| `df.isnull().values.any()` | Checks if **any missing value exists** in the DataFrame (similar to above).              |
| `df.isnull().values.sum()` | Total number of missing values in the DataFrame.                                         |


````python
df.isnull().sum()
````

- We can see that there are no missing values in the dataset.

# 11. Check with ASSERT statement 

To ensure data integrity before analysis or modelling, we can use assert statements to programmatically check for:

- Missing values <br/>
- Unexpected zero values (where not valid) <br/>
- Negative values (where not valid) <br/>

✦ Why Use Assert Statements?
- Confirms that data assumptions hold true <br/>
- Raises an AssertionError if conditions fail, preventing unnoticed errors later in your pipeline <br/>
- Provides confidence that the dataset is clean and analysis-ready <br/>

✦ How Assert Works

- If the condition is True, nothing happens.

- If the condition is False, an AssertionError is raised.



````python
#assert that there are no missing values in the dataframe

assert pd.notnull(df).all().all()
````

````python
#assert all values are greater than or equal to 0

assert (df >= 0).all().all()
````

- The above two commands do not throw any error. Hence, it is confirmed that there are no missing or negative values in the dataset.

- All the values are greater than or equal to zero.
